Вариант 6:
$$
\begin{cases}
F_1(x, y, \lambda) = x^4 + y^2 + 2\lambda xy - \lambda^4 - 1 = 0;\\
F_2(x, y, \lambda) = y - x - 4\lambda - 1 = 0
\end{cases},
$$
где при $\lambda = 0$ решение имеет вид $x = -1$, $y = 0$

Воспользуемся методом "продолжения по дуге".

Посчитаем производные:
$$
\begin{cases}
F_{1, x} (x, y, \lambda) = 4x^3 + 2\lambda y \\
F_{1, y} (x, y, \lambda) = 2y + 2\lambda x \\
F_{1, \lambda} (x, y, \lambda) = 2xy - 4\lambda^3 \\
F_{2, x} (x, y, \lambda) = -1 \\
F_{2, y} (x, y, \lambda) = 1 \\
F_{2, \lambda} (x, y, \lambda) = -4
\end{cases}
$$

Первой точкой нашей кривой будет $U_0 = (-1,\, 0,\, 0)$. Мы хотим найти вторую точку $U_1$, которая будет находится на малом расстоянии $\Delta s$ от $U_0$. Для начала построим касательный единичный вектор $(X_0,\, Y_0,\, \Lambda_0)$ из точки $U_0$, тогда получаем:
$$
\begin{cases}
F_{1, x} (U_0) X_0 + F_{1, y} (U_0) Y_0 + F_{1, \lambda} (U_0) \Lambda_0 = 0 \\
F_{2, x} (U_0) X_0 + F_{2, y} (U_0) Y_0 + F_{2, \lambda} (U_0) \Lambda_0 = 0 \\
X_0^2 + Y_0^2 + \Lambda_0^2 = 1
\end{cases}
$$

Немного схитрим и воспользуемся следующими заменами:
$$X_0 = \alpha X,\ \ \ \ \ Y_0 = \alpha Y,\ \ \ \ \ \Lambda_0 = \alpha = \pm \frac{1}{\sqrt{X^2 + Y^2 + 1}},$$
где знак при $\alpha$ определяет направление движения по нашей кривой.

Тогда для получения компонент первого касательного вектора достаточно подставить в формулы выше вычисленные значения $X$ и $Y$ из упрощенной системы:
$$
\begin{cases}
F_{1, x} (U_0) X + F_{1, y} (U_0) Y = -F_{1, \lambda} (U_0) \\
F_{2, x} (U_0) X + F_{2, y} (U_0) Y = -F_{2, \lambda} (U_0) 
\end{cases}
$$

Заметим, что из-за $\Delta s$ точки $U_{m + 1}$ и $U_m$ находятся на малом расстоянии друг от друга, а значит и их касательные будут иметь малый угол между собой. В таком случае условие единичности ($X_m^2 + Y_m^2 + \Lambda_m^2 = 1$) можно заменить приближенным:
$$X_mX_{m + 1} + Y_mY_{m + 1} + \Lambda_m\Lambda_{m + 1} = 1$$

Тогда система упрощается и нам не требуется выбирать направление движения по кривой, ведь это направление задается предыдущим вектором:
$$
\begin{cases}
F_{1, x} (U_{m + 1}) X_{m + 1} + F_{1, y} (U_{m + 1}) Y_{m + 1} + F_{1, \lambda} (U_{m + 1}) \Lambda_{m + 1} = 0 \\
F_{2, x} (U_{m + 1}) X_{m + 1} + F_{2, y} (U_{m + 1}) Y_{m + 1} + F_{2, \lambda} (U_{m + 1}) \Lambda_{m + 1} = 0 \\
X_mX_{m + 1} + Y_mY_{m + 1} + \Lambda_m\Lambda_{m + 1} = 1
\end{cases}
$$

Теперь вторая часть. Для получения непосредственно координат следующей точки кривой будем строить плоскость, для которой вектор $(X_m,\, Y_m,\, \Lambda_m)$ является нормальным, причем эта плоскость находится на расстоянии $\Delta s$ от точки $U_m$. Уравнение этой плоскости:
$$X_m(x - x_m) + Y_m(y - y_m) + \Lambda_m(\lambda - \lambda_m) - \Delta s = 0$$

Полагая $\Delta s$ малым, следующая точка $U_{m + 1} = (x_{m + 1},\, y_{m + 1},\, \lambda_{m + 1})$ окажется на этой плоскости. А значит помимо $F_1(U_{m + 1}) = 0$ и $F_2(U_{m + 1}) = 0$ ее координаты должны удовлетворять уравнению плоскости. Получается система из 3-х уравнений, которая однозначно задает точку $U_{m + 1}$ (для ее решения используем метод Ньютона, где первое приближение - точка пересечения касательной из $U_m$ с плоскостью).

Первое приближение:
$$x_{m + 1}^{(0)} = x_m + \Delta s \cdot X_m,\ \ \ \ \ y_{m + 1}^{(0)} = y_m + \Delta s \cdot Y_m,\ \ \ \ \ \lambda_{m + 1}^{(0)} = \lambda_m + \Delta s \cdot \Lambda_m$$

Сам итерационный процесс:
$$
\begin{pmatrix}
x_{m + 1}^{(k + 1)} \\ y_{m + 1}^{(k + 1)} \\ \lambda_{m + 1}^{(k + 1)}
\end{pmatrix} = 
\begin{pmatrix}
x_{m + 1}^{(k)} \\ y_{m + 1}^{(k)} \\ \lambda_{m + 1}^{(k)}
\end{pmatrix} -
\begin{pmatrix}
\Delta x^{(k + 1)} \\ \Delta y^{(k + 1)} \\ \Delta \lambda^{(k + 1)}
\end{pmatrix},
$$
где последний вектор находится из системы уравнений:

$$
\begin{pmatrix}
F_{1, x} (U_{m + 1}^{(k)}) & F_{1, y} (U_{m + 1}^{(k)}) & F_{1, \lambda} (U_{m + 1}^{(k)}) \\
F_{2, x} (U_{m + 1}^{(k)}) & F_{2, y} (U_{m + 1}^{(k)}) & F_{2, \lambda} (U_{m + 1}^{(k)}) \\
X_m & Y_m & \Lambda_m
\end{pmatrix}
\begin{pmatrix}
\Delta x^{(k + 1)} \\ \Delta y^{(k + 1)} \\ \Delta \lambda^{(k + 1)}
\end{pmatrix} =
\begin{pmatrix}
F_1 (U_{m + 1}^{(k)}) \\ F_2 (U_{m + 1}^{(k)}) \\ X_m(x_{m + 1}^{(k)} - x_m) + Y_m(y_{m + 1}^{(k)} - y_m) + \Lambda_m(\lambda_{m + 1}^{(k)} - \lambda_m) - \Delta s
\end{pmatrix}
$$

Посчитаю компоненты первого касательного вектора сам:
$$
\begin{cases}
F_{1, x} (-1,\, 0,\, 0) X + F_{1, y} (-1,\, 0,\, 0) Y = -F_{1, \lambda} (-1,\, 0,\, 0) \\
F_{2, x} (-1,\, 0,\, 0) X + F_{2, y} (-1,\, 0,\, 0) Y = -F_{2, \lambda} (-1,\, 0,\, 0) 
\end{cases} \implies
\begin{cases}
-4X = 0 \\
-X + Y = 4
\end{cases} \implies
\begin{cases}
X = 0 \\
Y = 4
\end{cases}
$$

$$\alpha = \pm \frac{1}{\sqrt{X^2 + Y^2 + 1}} = \pm \frac{1}{\sqrt{17}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
F_1 = lambda U: U[0][0] ** 4 + U[1][0] ** 2 + 2 * U[2][0] * U[0][0] * U[1][0] - U[2][0] ** 4 - 1
F_2 = lambda U: U[1][0] - U[0][0] - 4 * U[2][0] - 1
F_1x = lambda U: 4 * U[0][0] ** 3 + 2 * U[2][0] * U[1][0]
F_1y = lambda U: 2 * U[1][0] + 2 * U[2][0] * U[0][0]
F_1lm = lambda U: 2 * U[0][0] * U[1][0] - 4 * U[2][0] ** 3
F_2x = -1
F_2y = 1
F_2lm = -4

def next_point(U_m, tan, delta_s, eps = 10 ** -8):
    # first approximation
    U = U_m + delta_s * tan

    while True:
        L = np.array([[F_1x(U), F_1y(U), F_1lm(U)], [F_2x, F_2y, F_2lm], tan.T[0]])
        R = np.array([[F_1(U)], [F_2(U)], (tan.T @ (U - U_m))[0] - delta_s])
        delta = np.linalg.solve(L, R)
        U -= delta
        #print(delta)
        if np.amax(abs(delta)) < eps:
            break

    return U

def next_tangent(U, prev_tan):
    L = np.array([[F_1x(U), F_1y(U), F_1lm(U)], [F_2x, F_2y, F_2lm], prev_tan.T[0]])
    R = np.array([[0], [0], [1]])
    return np.linalg.solve(L, R)
```

```python
# bool
# true, if alpha is negative
neg_alpha = 0

delta_s = 10 ** -3

U = [np.array([[-1], [0], [0]])]
tan = [np.array([[0], [4 / np.sqrt(17)], [1 / np.sqrt(17)]])]
if neg_alpha:
    tan[-1] *= -1

while U[-1][2][0] * (-1) ** neg_alpha >= 0:
    U.append(next_point(U[-1], tan[-1], delta_s))
    tan.append(next_tangent(U[-1], tan[-1]))
```

```python
x = [U[i][0][0] for i in range(len(U))]
y = [U[i][1][0] for i in range(len(U))]
lm = [U[i][2][0] for i in range(len(U))]

if neg_alpha:
    bif_value = min(lm)
else:
    bif_value = max(lm)

print("Бифуркационное значение:", bif_value)


plt.plot([bif_value, bif_value], [min(x), max(x)], "k--")
plt.plot(lm, x)
plt.xlabel("lm")
plt.ylabel("x")
plt.grid(True)

plt.figure()
plt.plot([bif_value, bif_value], [min(y), max(y)], "k--")
plt.plot(lm, y)
plt.xlabel("lm")
plt.ylabel("y")
plt.grid(True)

fig = plt.figure('Cool curve')
ax = plt.axes(projection='3d')
ax.plot3D(x, y, lm);
```
